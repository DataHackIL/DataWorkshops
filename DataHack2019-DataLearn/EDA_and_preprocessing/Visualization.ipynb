# Scatter plots

```python
x = np.random.normal(0, 1, 1000)
y = x + np.random.normal(0, 0.2, 1000)
```

```python
# basic scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(x, y, c='green', s=50, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

## Histograms

```python
plt.hist(x, color='firebrick', edgecolor='k')
plt.xlabel('x')
plt.ylabel('count')
plt.grid(True)
plt.xlim(-4, 4)
plt.title('Distribution of x');
```

```python
# cumulative histogram
plt.hist(x, bins=10, range=(-2, 2), cumulative=True, color='firebrick', edgecolor='k')
plt.xlabel('x')
plt.ylabel('count')
plt.grid(True)
plt.title('Cumulative values of x');
```

```python
# stacked historgrams
plt.hist([x, y], color=('firebrick', 'royalblue'),
         histtype='bar', edgecolor='k', label=('x', 'y'))
plt.xlabel('x')
plt.ylabel('count')
plt.grid(True)
plt.legend()
plt.title('Distribution of x and y');
```

## Box plots

```python
plt.boxplot(x)
plt.grid(True)
plt.title('Distribution of x');
```

```python
plt.boxplot([x, y], showfliers=False)
plt.grid(True)
plt.title('Distribution of x and y')
plt.xticks([1, 2], ['x', 'y'], fontsize=12);
```

### Subplots

```python
plt.figure(figsize=(8, 8))

# first subplot
plt.subplot(2, 2, 1)
plt.scatter(x, y, marker='o', s=32, edgecolor='k')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('y vs. x')

# second subplot
plt.subplot(2, 2, 2)
plt.plot(x, c='firebrick', linewidth=1, alpha=0.7, label='x')
plt.plot(y, c='royalblue', linewidth=1, alpha=0.7, label='y')
plt.xlabel('x, y')
plt.grid(True)
plt.legend()
plt.title('x and y')

# third subplot
plt.subplot(2, 2, 3)
plt.hist(x, bins=10, color='firebrick', edgecolor='k')
plt.xlabel('x')
plt.ylabel('count')
plt.grid(True)
plt.title('Distribution of x')

# fourth subplot
plt.subplot(2, 2, 4)
plt.hist(y, bins=10, color='royalblue', edgecolor='k')
plt.xlabel('y')
plt.ylabel('count')
plt.grid(True)
plt.title('Distribution of y')

plt.tight_layout()
```
