### Indexing - loc vs. iloc

The main 2 options to refer to a specific item in our data:
* index based (loc)
* location based (iloc)

Pay attention to which one of the above you are using - sometime it can be critical.

```python
# The difference between index based and location based data get
print(s.loc[2])
print(s.iloc[2])
```
